```python
img = Image.open("potato-bug.jpg")

# Convert to grayscale
gray_img = img.convert("L")

# Convert to numpy array
gray_arr = np.array(gray_img)

# Flatten the image (2D  1D)
flat_img = gray_arr.ravel()

# Plot histograms
plt.figure(figsize=(12, 5))

# Histogram of original grayscale image (2D array visualization)
plt.subplot(1, 2, 1)
plt.hist(gray_arr.flatten(), bins=256, color="blue", alpha=0.7)
plt.title("Histogram of Grayscale Image")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")

# Histogram of flattened image (1D version)
plt.subplot(1, 2, 2)
plt.hist(flat_img, bins=256, color="green", alpha=0.7)
plt.title("Histogram of Flattened Image")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
img = Image.open("potato-bug.jpg")

# Convert to numpy array (RGB)
img_arr = np.array(img)

# Convert to grayscale
gray_img = img.convert("L")
gray_arr = np.array(gray_img)

# Flatten grayscale image
flat_img = gray_arr.ravel()

plt.figure(figsize=(15, 5))

# Histogram of Original RGB image (per channel)
plt.subplot(1, 3, 1)
colors = ("red", "green", "blue")
for i, color in enumerate(colors):
    plt.hist(img_arr[:, :, i].ravel(), bins=256, color=color, alpha=0.5, label=f"{color.upper()} channel")
plt.title("Histogram of Original Color Image")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")
plt.legend()

# Histogram of Grayscale image
plt.subplot(1, 3, 2)
plt.hist(gray_arr.flatten(), bins=256, color="gray", alpha=0.7)
plt.title("Histogram of Grayscale Image")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")

# Histogram of Flattened grayscale image
plt.subplot(1, 3, 3)
plt.hist(flat_img, bins=256, color="black", alpha=0.7)
plt.title("Histogram of Flattened Grayscale Image")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# . Contrast Adjustment
plt.subplot(3, 3, 8)
plt.imshow(gray_arr, cmap="gray", clim=(0, 175))
plt.title("Contrast Adjusted (clim 0175)")
plt.axis("off")
```

```python
# 9. Interpolation Comparison (Zoomed)
plt.subplot(3, 3, 9)
plt.imshow(gray_arr[100:200, 200:300], cmap="gray", interpolation="bicubic")
plt.title("Interpolation: Bicubic (Zoomed)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
from scipy.ndimage import sobel

edge_x = sobel(gray_arr, axis=0)
edge_y = sobel(gray_arr, axis=1)
edges = np.hypot(edge_x, edge_y)

plt.imshow(edges, cmap="gray")
plt.title("Edges (Sobel Filter)")
plt.axis("off")
plt.show()
```

```python
hist, bins = np.histogram(gray_arr.flatten(), bins=256, range=[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

equalized = np.interp(gray_arr.flatten(), bins[:-1], cdf_normalized)
equalized_img = equalized.reshape(gray_arr.shape)

plt.imshow(equalized_img, cmap="gray")
plt.title("Histogram Equalized Image")
plt.axis("off")
plt.show()
```

```python
noise = np.random.randint(0, 50, gray_arr.shape, dtype='uint8')
noisy_img = np.clip(gray_arr + noise, 0, 255)

plt.imshow(noisy_img, cmap="gray")
plt.title("Image with Random Noise")
plt.axis("off")
plt.show()
```
